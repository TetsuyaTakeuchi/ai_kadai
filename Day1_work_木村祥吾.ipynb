目的：ガソリンごとの燃費(ℓ/100km:100km)を推測する。

ガソリンごとなのでモデルは２つ必要。
データ形式はcsv。

表示前にデータをざっと確認。
specials列は天候の内容でAC,rain,sunに反映されているため除外。(snowが気になる)
refill liters,refill gas列はほぼ空データだが消費量と組み合わせれば車体重量に関係がありそう。


```python
%matplotlib inline
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D #3D散布図の描画
```

```python
#データを取得：使わないとこはあらかじめ取らない
df_csv = pd.read_csv("./csv/measurements.csv", usecols=lambda x: x not in ['specials', 'refill liters', 'refill gas'])
df_csv
```

```python
#型チェック
df_csv.dtypes
```

```python
#文字列を数値にする
#欠損チェック
df_csv.isnull().sum()
```

```python
#少ないので削除
df_csv.dropna(inplace=True)
```

```python
#文字列を数値にするため変換
df_csv['distance'] = df_csv['distance'].str.replace(',', '.')
df_csv['consume'] = df_csv['consume'].str.replace(',', '.')
df_csv['temp_inside'] = df_csv['temp_inside'].str.replace(',', '.')
```

```python
#数値変換：破壊的に追加(dfが変更される)。
df_csv['distance_float'] = df_csv['distance'].astype(np.float64)
df_csv['consume_float'] = df_csv['consume'].astype(np.float64)
df_csv['temp_inside_float'] = df_csv['temp_inside'].astype(np.float64)
```

```python
#csvのgas_typeでcsvを２つに分割。
df_e10 = df_csv[df_csv['gas_type']== 'E10']
df_sp80 =  df_csv[df_csv['gas_type'] == 'SP98']
```

```python
#目的は目的：ガソリンごとの燃費(ℓ/100km:100km)を推測する。
#よって目的変数はconsume_float
#目的変数の説明となる変数が説明変数なので相関関係にある変数を選択する。
#が、すべての変数が低い相関・相関なしの範囲なのですべての変数を用いる。
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# 係数を求める
y = df_e10["consume_float"].values
X = df_e10[['speed', 'temp_outside', 'AC', 'rain', 'sun', 'distance_float','temp_inside_float']].values
e10_regr = LinearRegression(fit_intercept=True)
e10_regr.fit(X, y)
```

```python
#性能を評価する
# 値を予測
y_pred = e10_regr.predict(X)

# MSEを計算
mse = mean_squared_error(y, y_pred) 
print("MSE = %s"%round(mse,3) )  

# MAEを計算
mae = mean_absolute_error(y, y_pred) 
print("MAE = %s"%round(mae,3) )

# RMSEを計算
rmse = np.sqrt(mse)
print("RMSE = %s"%round(rmse, 3) )
```

```python
# 係数を求める
y = df_sp80["consume_float"].values
X = df_sp80[['speed', 'temp_outside', 'AC', 'rain', 'sun', 'distance_float','temp_inside_float']].values
sp80_regr = LinearRegression(fit_intercept=True)
sp80_regr.fit(X, y)
```

```python
#性能を評価する
# 値を予測
y_pred = sp80_regr.predict(X)

# MSEを計算(誤差の2乗の平均)
mse = mean_squared_error(y, y_pred) 
print("MSE = %s"%round(mse,3) )  

# MAEを計算:誤差の絶対値の平均
mae = mean_absolute_error(y, y_pred) 
print("MAE = %s"%round(mae,3) )

# RMSEを計算:誤差の絶対値を２乗した平均＝大きな誤差に厳しい
rmse = np.sqrt(mse)
print("RMSE = %s"%round(rmse, 3) )
```
